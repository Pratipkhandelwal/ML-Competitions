```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import seaborn as sns
```

```python
train = pd.read_csv('criminal_train.csv')
test = pd.read_csv('criminal_test.csv')
```

```python
Y = train['Criminal']
train = train.drop(['PERID','Criminal'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train.head()
```

```python
X = pd.DataFrame(train)
y = Y

X_train, X_test, Y_train, Y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,Y_train)
```

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix

print(confusion_matrix(Y_test,predictions))
print(classification_report(Y_test,predictions))
```

```python
from sklearn.ensemble import RandomForestClassifier as RFC
```

```python
rfc = RFC(n_estimators=200)
```

```python
rfc.fit(X_train,Y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
print(confusion_matrix(Y_test,rfc_pred))
print(classification_report(Y_test,rfc_pred))
```

```python
Z = test['PERID']
```

```python
test = test.drop(['PERID'],axis=1)
```

```python
predict_final = rfc.predict(test)
```

```python
predict_final
```

```python
submiss = pd.DataFrame()
submiss['PERID'] = Z

submiss['Criminal'] = predict_final
submiss.head()
```
